```python
#single number collatz conjecture
n=10
count=0
while n!=1:
    if n%2 == 0:
        n=n/2
        print(n)
        count=count +1
    else:
        n=(n*3)+1
        print(n)
        count=count+1
print("Iterations= {}".format(count))
```

```python
#multiple number collatz conjecture
narray=arange(2,100000,1)
countnum=[]
for i in narray:
    n=i
    count=0
    while n!=1:
        if n%2 == 0:
            n=n/2
            count=count +1
        else:
            n=(n*3)+1
            count=count+1
    countnum.append(count)
```

```python
plot(narray,countnum)
title("Collatz Conjecture Iterations")
xlabel("Integer")
ylabel("Iterations to reach 1")
```
